```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('EURUSD_D1 (1).csv')
```

```python
df.head()
```

```python
line = df.columns
line
```

```python
# Add column names to the DataFrame
df.columns = ['Time', 'Open', 'High', 'Low', 'Close', 'Volume']
```

```python
df
```

```python
new_line = pd.DataFrame({'Time': [line[0]], 'Open': [line[1]], 'High': [
                        line[2]], 'Low': [line[3]], 'Close': [line[4]], 'Volume': [line[5]]})


df = pd.concat([new_line, df]).reset_index(drop=True)
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Time'], format='%Y-%m-%d ')
df
```

```python
df['Day'] = df['Date'].dt.day_name()
df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = df['Date'].dt.quarter
df['Year'] = df['Date'].dt.year
df
```

```python
df.dtypes
```

```python
df['Tomorrow'] = df['Close'].shift(-1)
df
```

```python
cols = df[['Open', 'High', 'Low', 'Close', 'Volume', 'Tomorrow']]
cols
```

```python
# convert the columns to numeric
for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.dtypes
```

```python
df['Target'] = (df['Tomorrow'] > df['Close']).astype(int)
```

```python
df['Moving Average'] = df['Close'].rolling(window=20).mean()
df['MACD'] = df['Close'].ewm(span=12, adjust=False).mean(
) - df['Close'].ewm(span=26, adjust=False).mean()
df['Signal Line'] = df['MACD'].ewm(span=9, adjust=False).mean()
df['RSI'] = 100 - (100 / (1 + (df['Close'].diff().fillna(0).rolling(
    window=14).apply(lambda x: x[x > 0].mean() / -x[x < 0].mean()))))

df
```

```python
df = df.dropna().reset_index(drop=True)
df
```

```python
df['Day'].value_counts()
```

```python
# drop all lines where the day is Sunday
df = df[df['Day'] != 'Sunday'].reset_index(drop=True)
```

```python
# label encoding for Day, Month
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Day_enc'] = le.fit_transform(df['Day'])
df['Month_enc'] = le.fit_transform(df['Month'])
```

```python
df['Day_enc'].unique()
```

```python
df['Month_enc'].unique()
```

```python
df
```

```python
from sklearn.ensemble import RandomForestClassifier

model = model = RandomForestClassifier(
    n_estimators=100, min_samples_split=100, random_state=1)
train = df.iloc[:-100]
test = df.iloc[-100:]

predictors = ['Open', 'High', 'Low', 'Close', 'Volume', 'Moving Average',
              'MACD', 'Signal Line', 'RSI', 'Day_enc', 'Month_enc', 'Year', 'Quarter']
target = 'Target'

model.fit(train[predictors], train[target])
```

```python
from sklearn.metrics import precision_score, classification_report, confusion_matrix
preds = model.predict(test[predictors])
preds = pd.Series(preds, index=test.index)
```

```python
from sklearn.metrics import classification_report, confusion_matrix


precision = precision_score(test["Target"], preds)
classification_report = classification_report(test["Target"], preds)
confusion_matrix = confusion_matrix(test["Target"], preds)

print('Precision:', precision)
print('Classification Report:\n', classification_report)
print('Confusion Matrix:\n', confusion_matrix)
```

```python
horizons = [2, 5, 60, 250, 1000]
new_predictors = []
for horizon in horizons:
    rolling_average = df.rolling(horizon).mean()
    ratio_column = f"Close_Ratio_{horizon}"
    df[ratio_column] = df["Close"] / rolling_average["Close"]
    trend_column = f"Trend_{horizon}"
    df[trend_column] = df.shift(1).rolling(horizon).sum()["Target"]
    new_predictors += [ratio_column, trend_column]
```

```python
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier, XGBRFClassifier
```

```python
model_name = []
precision = []
models = [
    RandomForestClassifier(
        n_estimators=100, min_samples_split=100, random_state=1),
    GradientBoostingClassifier(
        n_estimators=100, min_samples_split=100, random_state=1),
    RandomForestClassifier(n_estimators=200, min_samples_split=5,
                           min_samples_leaf=2, max_depth=20, bootstrap=True, random_state=1),
    ExtraTreesClassifier(
        n_estimators=100, min_samples_split=100, random_state=1),
    AdaBoostClassifier(n_estimators=100, random_state=1),
    DecisionTreeClassifier(random_state=1),
    XGBClassifier(objective='binary:logistic', random_state=1),
    XGBRFClassifier(kwds={'objective': 'binary:logistic'}, random_state=1),
]

for model in models:
    model.fit(train[predictors], train[target])
    preds = model.predict(test[predictors])
    precisionScore = precision_score(test["Target"], preds)
    model_name.append(model)
    precision.append(precisionScore)
```

```python
models_df = pd.DataFrame(
    {"Model Name": model_name, "Precision": precision, "Accuracy": accuracy})
models_df = models_df.astype(
    {"Model Name": str, "Precision": float, "Accuracy": float})
models_df = models_df.sort_values(by="Precision", ascending=False)

best_model = models_df.iloc[0]
best_model_name = best_model["Model Name"]
print(f"Best Model is {best_model_name} with Precision of {best_model['Precision']}")
```
